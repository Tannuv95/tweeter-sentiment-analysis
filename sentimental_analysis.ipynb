```python
import pandas as pd
```

```python
stock_price_data=pd.read_csv("stock_prices.csv")
stock_price_data
```

```python
tweeter_sentiment_data=pd.read_csv("tweeter _sentiments.csv")
tweeter_sentiment_data
```

```python
 tweeter_sentiment_data.info()
```

```python
tweeter_sentiment_data=tweeter_sentiment_data.dropna(axis=1)
tweeter_sentiment_data
```

```python
stock_price_data=stock_price_data.dropna(axis=1)
stock_price_data
```

```python
tweeter_sentiment_data.rename(columns={'date':'Date'},inplace=True)
```

```python
merged_dataframe=stock_price_data.merge(tweeter_sentiment_data,on="Date")
merged_dataframe
```

```python
merged_dataframe.to_csv("merged_stock_tweeter_dataframe.csv",index=False)
```

```python
dataframe=merged_dataframe[["Date","Adj Close","Volume","ts_polarity","twitter_volume"]]
dataframe
```

```python
dataframe.set_index("Date",inplace=True)
```

```python
dataframe
```

```python
POSITIVE_SENTIMENT_THRESHOLD =0.05
NEGATIVE_SENTIMENT_THRESHOLD=-0.05
sentiments=[]
```

```python
for sentiment_score in dataframe["ts_polarity"]:
  if sentiment_score>=POSITIVE_SENTIMENT_THRESHOLD:
    sentiments.append("Postive")
  elif sentiment_score<=NEGATIVE_SENTIMENT_THRESHOLD:
      sentiments.append("Negative")
  else:
    sentiments.append("Neutral")


dataframe["Sentiment"]=sentiments
```

```python
dataframe
```

```python
dataframe["Sentiment"].value_counts()
```

```python
dataframe["Adj Close"].value_counts()
```

```python
dataframe["Price difference"]= dataframe["Adj Close"].diff()
dataframe.dropna(inplace=True)
dataframe
```

```python
import numpy as np
RISE=1
FALL=0
dataframe["Stock Trend"]=np.where(
    dataframe["Price difference"]> 0,RISE,FALL
)
```

```python
dataframe
```

```python
new_dataframe=dataframe[["Adj Close","Volume","twitter_volume","Sentiment","Stock Trend"]]
new_dataframe=pd.get_dummies(new_dataframe,columns=["Sentiment"])
new_dataframe
```

```python
X=new_dataframe.copy()
X.drop("Stock Trend",axis =1,inplace=True)
```

```python
X
```

```python
Y=new_dataframe["Stock Trend"].values.reshape(-1,1)
Y[:10]
```

```python
SPLIT=int(0.8*len(X))
X_train=X[:SPLIT]
X_test=X[SPLIT:]

Y_train=Y[:SPLIT]
Y_test=Y[SPLIT:]
```

```python
Y_test
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaler=scaler.fit(X_train)
X_train_scaled=X_scaler.transform(X_train)
X_test_scaled=X_scaler.transform(X_test)
```
